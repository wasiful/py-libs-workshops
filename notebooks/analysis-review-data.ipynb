```python
import json
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import gzip
```

```python

data_path = Path("../data").joinpath("amzn_review", "Industrial_and_Scientific.jsonl.gz")
metadata_path = Path("../data").joinpath("amzn_review", "meta_Industrial_and_Scientific.jsonl.gz")
        
```

```python
def load_json_into_dataframe(file: Path, sample=False) -> pd.DataFrame:
    data = []
    count = 0
    with gzip.open(file, 'r') as fp:
        for line in tqdm(fp):
            data_item = json.loads(line.strip())
            data.append(data_item)
            count += 1
            if sample and count == 100000:
                break
    return pd.DataFrame(data)
```

```python
data_df = load_json_into_dataframe(data_path, sample=False)
```
